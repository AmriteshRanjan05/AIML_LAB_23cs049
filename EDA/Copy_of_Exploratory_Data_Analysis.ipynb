# [Objective 03](#pandas-visualizations) Describe and discriminate between basic data types

## Overview

One of the cornerstones of Exploratory Data Analysis (EDA) is being able to identify variable types.  We will need different statistical methods to display and describe each of these different types of data.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (you already have column names from the URL or dataset file)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/flags/flag.data"

# If column headers aren't predefined, you need to set them first:
column_headers = [
    "name", "landmass", "zone", "area", "population", "language", "religion", "bars",
    "stripes", "colours", "red", "green", "blue", "gold", "white", "black", "orange",
    "mainhue", "circles", "crosses", "saltires", "quarters", "sunstars", "crescent",
    "triangle", "icon", "animate", "text", "topleft", "botright"
]

dataset = pd.read_csv(url, names=column_headers)

print(dataset.shape)
print(dataset.head())
```

```
print(dataset.info())
print(dataset.dtypes)
```

```
print(dataset.isna().sum())
print(f"Total Missing Values: {dataset.isna().sum().sum()}")
```

```
print(dataset.describe())
print(dataset.describe(include=['object']))
```

```
numerical_cols=dataset.select_dtypes(include=['int64', 'float64']).columns
categorical_cols=dataset.select_dtypes(include=['object']).columns

print("Numerical Columns:", list(numerical_cols))
print("Categorical Columns:", list(categorical_cols))
```

```
dataset[numerical_cols].hist(figsize=(15, 10), bins=30, color='orange', edgecolor='white')
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```
for feature in categorical_cols:
    plt.figure(figsize=(12,7))
    sns.countplot(data=dataset, x=feature, palette="Set1")
    plt.title(f"Distribution of {feature}")
    plt.xticks(rotation=90)
    plt.show()
```

```
plt.figure(figsize=(20,15))
sns.heatmap(dataset[numerical_cols].corr(),annot=True,cmap="viridis",fmt=".3f")
plt.title("Correlation Heatmap for Numerical Features")
plt.show()
```

```
plt.figure(figsize=(10,8))
sns.boxplot(data=dataset,x="landmass",y="area",palette="bright")
plt.title("Area Distribution by Landmass")
plt.show()
```

```
plt.figure(figsize=(18,10))
sns.boxplot(data=dataset[numerical_cols])
plt.xticks(rotation=90)
plt.title("Outlier Detection in Numerical Variables")
plt.show()
```
